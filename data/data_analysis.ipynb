```python
import json
import pandas as pd
import numpy as np
import os
import csv
```

```python
test_temporalspatial = json.load(open('./test_all_json/test_temporalspatial.json', encoding="utf-8"))
test_temporalspatial_pd = pd.DataFrame(test_temporalspatial)
test_temporalspatial_pd
```

```python
## ensemble_three_methods
test_json = json.load(open('./KFold_catboost.json', encoding="utf-8"))
test_resuts = [sample['popularity_score'] for sample in test_json['result']]
# test_resuts
```

```python
test_score = pd.DataFrame(test_resuts)
test_score.columns = ['popularity']
# test_score
```

```python
test_score_tem = pd.concat([test_temporalspatial_pd, test_score], axis = 1)
test_score_tem
```

```python
sorted_test = test_score_tem.sort_values(by="Postdate", ascending=True) 
sorted_test
```

```python
import datetime
sorted_test['datetime'] = sorted_test['Postdate'].apply(lambda x: str(datetime.datetime.fromtimestamp(int(x)).year) + str(datetime.datetime.fromtimestamp(int(x)).month)+str(datetime.datetime.fromtimestamp(int(x)).day))
# sorted_test['datetime']
```

```python
sorted_test
```

```python
reindexed_sorted_test = sorted_test.reset_index()
reindexed_sorted_test
```

```python
## 计算每一天预测结果的值
from tqdm import trange
all_mean_norm = []
num = 0
# popu = 8.8814
popu = 0
for index in trange(180580):
  
    if reindexed_sorted_test['datetime'][index] == reindexed_sorted_test['datetime'][index + 1]:
        if index == 180579:
            popu = popu + reindexed_sorted_test['popularity'][index] + reindexed_sorted_test['popularity'][index + 1]
            num = num + 2
            mean_popu = popu / num
            all_mean_norm.append(mean_popu)
            
        else:
            popu = popu + reindexed_sorted_test['popularity'][index]
            num = num + 1
    else:
        popu = popu + reindexed_sorted_test['popularity'][index]
        num = num + 1
        mean_popu = popu / num
        all_mean_norm.append(mean_popu)
        popu = 0
        num = 0
        
all_mean_norm
```

```python
print('mean_both_datasets:', np.mean(reindexed_sorted_test['popularity']))
print('std_both_datasets:', np.std(reindexed_sorted_test['popularity']))
```
